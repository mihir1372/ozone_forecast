```python
# Define the split point
split_point = int(len(data1) * 0.8)
```

```python
# Split into training and test sets
train_data = data1.iloc[:split_point]
test_data = data1.iloc[split_point:]
```

```python
# Plot the split data
plt.figure(figsize=(8, 6))
plt.plot(train_data.index, train_data['Daily Max 8-hour Ozone Concentration'], label='Training Data')
plt.plot(test_data.index, test_data['Daily Max 8-hour Ozone Concentration'], label='Testing Data')
plt.title('Ozone Concentration - Train and Test Split')
plt.xlabel('Date')
plt.ylabel('Ozone Concentration (ppm)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# # Assign the frequency (assuming daily data; change if different)
# data1 = data1.asfreq('D')

# # Verify that the index has a frequency now
# print(data1.index.freq)
```

```python
# Initialize and fit the model on the training data
esm_model = ExponentialSmoothing(
    train_data['Daily Max 8-hour Ozone Concentration'],
    trend='add',                # Capture trend (linear trend in this case)
    seasonal='add',              # Capture seasonality (additive)
    seasonal_periods=365         # Assume yearly seasonality
)
esm_fit = esm_model.fit()
```

```python
# Forecast for the length of the test data
forecast = esm_fit.forecast(steps=len(test_data))

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data['Daily Max 8-hour Ozone Concentration'], label='Training Data')
plt.plot(test_data.index, test_data['Daily Max 8-hour Ozone Concentration'], label='Actual Test Data')
plt.plot(test_data.index, forecast, label='Forecast', color='red')
plt.title('Ozone Concentration Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Ozone Concentration (ppm)')
plt.legend()
plt.show()
```

```python
# Calculate error metrics
mae = mean_absolute_error(test_data['Daily Max 8-hour Ozone Concentration'], forecast)
mse = mean_squared_error(test_data['Daily Max 8-hour Ozone Concentration'], forecast)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

```python
# Convert 'Date' to datetime for proper time-series handling
data1['Date'] = pd.to_datetime(data1['Date'])
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(data1.corr(), annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Plot 2: Time Series Plot for Ozone Concentration
plt.figure(figsize=(12, 6))
plt.plot(data1['Date'], data1['Daily Max 8-hour Ozone Concentration'], label='Ozone Concentration')
plt.title("Time Series of Ozone Concentration")
plt.xlabel("Date")
plt.ylabel("Ozone Concentration (ppm)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot 3: Monthly Trends in Ozone Concentration
data1['Month'] = data1['Date'].dt.month
monthly_avg = data1.groupby('Month')['Daily Max 8-hour Ozone Concentration'].mean()

plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg, marker='o', linestyle='-', label='Monthly Average')
plt.title("Monthly Trends in Ozone Concentration")
plt.xlabel("Month")
plt.ylabel("Average Ozone Concentration (ppm)")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid()
plt.legend()
plt.show()
```

```python
# Plot 4: Boxplot of Ozone Concentration by Month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Daily Max 8-hour Ozone Concentration', data=data1, palette='coolwarm')
plt.title("Boxplot of Ozone Concentration by Month")
plt.xlabel("Month")
plt.ylabel("Ozone Concentration (ppm)")
plt.show()
```
